```python
import pandas as pd
```

```python
#Firstly in smsspamcollection read from file
data = pd.read_csv('smsspamcollection/SMSSpamCollection', delimiter='\t',header=None)
```

```python
# Data includes two column. First column show that the message is spam or not
# Second column shows the message

data.head()
```

```python
# First column of data is y. 
# Second column of data is X.
X=data[1]
y=data[0]
```

```python
# we should split data as train and split to avoid overfitting
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

```python
# In this step our text messages convertted to float points.
# We should vectorize the text data to classify
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
X_train=vectorizer.fit_transform(X_train)
X_test=vectorizer.transform(X_test)
```

```python
# Logistic regression is used to classify data
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()

model.fit(X_train,y_train)
```

```python
# Predict our models
predictions=model.predict(X_test)
```

```python
# Compare all predictions its true values
for x in range(len(predictions)):
    print(predictions[x],y_test.iloc[x])
```

```python
# Now, we can see classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
# precision means, lets say if I predict 100 times which is spam, 1.00 means all predicton is spam.
# recall means, let say We have 100 spam message, I correctly predicted %78 of them
```
